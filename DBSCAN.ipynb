```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
```

```python
dt = pd.read_csv('https://github.com/ezioauditore-tech/AI/raw/main/datasets/Mall_Customers.csv')
dt
```

```python
fet=["Annual Income (k$)","Spending Score (1-100)"]
x=dt[fet]
```

```python
xstd=StandardScaler().fit_transform(x)
```

```python
dbscan=DBSCAN(eps=0.5,min_samples=4)
clus=dbscan.fit_predict(xstd)
clus
```

```python
centroids=pd.DataFrame(xstd).groupby(clus).mean().values
```

```python
# prompt: plot the all the values in graph

plt.scatter(xstd[:,0],xstd[:,1],c=clus,cmap='rainbow')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('DBSCAN Clustering')
plt.show()
```
